```python
df_US['country'] = 'US'
df_IN['country'] = 'IN'

df = pd.concat([df_US, df_IN], ignore_index=True)
```

```python
df.head()
```

```python
df = df.drop(columns=['thumbnail_link', 'comments_disabled', 'ratings_disabled', 'video_error_or_removed'])
```

Deleted the unneccesary columns, not needed for content reference

```python
df['description'] = df['description'].fillna('')
```

```python
df['description'].isnull().sum()
```

```python
df['content'] = (
    df['title'].fillna('') + ' ' +
    df['tags'].fillna('') + ' ' +
    df['description'].fillna('')
)
```

```python
df.head()
```

```python
df['content']
```

Replacing separator tags with spaces for TF - IDF vectorizer

```python
df['content'] = df['content'].str.replace('|', ' ', regex=False)
```

```python
print(df['content'].iloc[4][:300])
```

```python
df = df.drop_duplicates(subset='video_id', keep='first')

print(df.shape)
print(df['video_id'].nunique())
```

Finally, we have a total of 22427 unique videos

```python
df.to_csv('data/cleaned_videos.csv', index=False)
print("Saved!")
```
